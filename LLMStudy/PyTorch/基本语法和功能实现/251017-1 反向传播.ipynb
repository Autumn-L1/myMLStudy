3步
	1.前向传播：计算损失
	2.计算局部梯度
	3.反向传播：计算d_Loss/d_Weight

一个例子![[Pasted image 20251017110941.png]]

```
import torch
x = torch.tensor(1.0)
y = torch.tensor(2.0)
w = torch.tensor(1.0, requires_grad = True)

# FP
y_hat = w*x
loss = (y_hat - y)**2

print(loss)

#BP
loss.backward()
print(w.grad)

# update weight
# more iters...
```
